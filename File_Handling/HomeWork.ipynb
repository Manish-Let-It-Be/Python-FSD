```python
#1 WAP to find the average of all the numbers stored in a file.
def write(filename):
    try: 
        with open(filename, 'w') as file:
            while True:
                data = input("Input numbers, 'done' to finish: ")
                if data.lower() == 'done':
                    break
                numbers = float(data)
                file.write(f"{numbers}\n")
    except Exception as e:
        print(e.args)

def average(filename):
    try:
        with open(filename, 'r') as file:
            data = file.read()
            numbers = data.splitlines()
            total = sum(float(num) for num in numbers)
            count = len(numbers)
            if count == 0:
                print("No number!")
            else:
                average =  total / count
                print(f"Average: {average}")
    except Exception as e:
        print(e.args)

if __name__ == "__main__":
    filename = "numbers.txt"
    write(filename)
    average(filename)
```

```python
#2 WAP to remove duplicate lines from a file.
def write(filename, lines):
    try:
        with open(filename, 'w') as file:
            file.writelines(lines)
    except Exception as e:
        print(e.args)

def remove_duplicate(filename):
    try:
        with open(filename, 'r') as file:
            lines = file.readlines()
            unique_lines = set(lines)
            write(filename, unique_lines)
    except Exception as e:
        print(e.args)

if __name__ == "__main__":
    filename = input("Enter the file name: ")
    lines = []

    while True:
        line = input("Enter a line, 'done' to finish: ")
        if line.lower() == 'done':
            break
        lines.append(line.strip() + '\n')
    write(filename, lines)
    remove_duplicate(filename)
```

```python
#3 WAP to replace the found word with a new word.

def replace_word_in_file(filename, old_word, new_word):
    try:
        with open(filename, 'r') as file:
            content = file.read()
        
        updated_content = content.replace(old_word, new_word)
        
        with open(filename, 'w') as file:
            file.write(updated_content)
        
    except Exception as e:
        print(e.args)

def display_file_content(filename):
    try:
        with open(filename, 'r') as file:
            content = file.read()
            print(content)
    except Exception as e:
        print(e.args)

if __name__ == "__main__":
    filename = input("Enter the file name: ")
    
    print("Original content of the file:")
    display_file_content(filename)

    old_word = input("Enter the word to replace: ")
    new_word = input("Enter the new word: ")

    replace_word_in_file(filename, old_word, new_word)

    print("Updated content of the file:")
    display_file_content(filename)
```

```python
#4 WAP to remove blank lines from the file.

def remove_blank_lines_from_file(filename):
    try:
        with open(filename, 'r') as file:
            lines = file.readlines()
        
        non_blank_lines = [line for line in lines if line.strip() != '']
        
        with open(filename, 'w') as file:
            file.writelines(non_blank_lines)
        
    except Exception as e:
        print(e.args)

def display_file_content(filename):
    try:
        with open(filename, 'r') as file:
            content = file.read()
            print(content)
    except Exception as e:
        print(e.args)

if __name__ == "__main__":
    filename = input("Enter the file name: ")
    
    print("Original content of the file:")
    display_file_content(filename)

    remove_blank_lines_from_file(filename)

    print("Updated content of the file:")
    display_file_content(filename)
```

```python
#5 WAP to find the count of vowels and consonants in a given file.
def count_vowels_and_consonants(filename):
    vowels = "aeiouAEIOU"
    consonant_count = 0
    vowel_count = 0

    try:
        with open(filename, 'r') as file:
            content = file.read()
        
        for char in content:
            if char.isalpha():  
                if char in vowels:
                    vowel_count += 1
                else:
                    consonant_count += 1
        
        print(f"Total Vowels: {vowel_count}")
        print(f"Total Consonants: {consonant_count}")
    
    except Exception as e:
        print(e.args)

def display_file_content(filename):
    try:
        with open(filename, 'r') as file:
            content = file.read()
            print(content)
    except Exception as e:
        print(e.args)

if __name__ == "__main__":
    filename = input("Enter the file name: ")
    
    print("Original content of the file:")
    display_file_content(filename)

    count_vowels_and_consonants(filename)
```
